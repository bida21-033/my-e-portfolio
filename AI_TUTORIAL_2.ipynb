```python
import matplotlib.pyplot as plt
import numpy as np

# Create an array of input values (sensor readings)
x_values = np.linspace(0, 10, 100)
y_values = np.array([1 if (0.7 * x - 3.5) >= 0 else 0 for x in x_values])

# Plot the sensor readings and the decision boundary
plt.plot(x_values, y_values, label="Decision Boundary")
plt.axvline(x=5, color='r', linestyle='--', label="Decision Boundary (x = 5)")
plt.xlabel("Sensor Reading (x)")
plt.ylabel("Output (y)")
plt.title("Decision Boundary for Single Layer Perceptron")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt
import numpy as np

# Assuming x_values and y_values are defined from previous code

# Create and train the MLP classifier
mlp = MLPClassifier(hidden_layer_sizes=(10,), max_iter=1000) # Example parameters
mlp.fit(x_values.reshape(-1, 1), y_values) # Reshape x_values for sklearn

# Now you can plot the loss curve
plt.plot(mlp.loss_curve_)
plt.title('MLP Loss Curve')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.show()
```
